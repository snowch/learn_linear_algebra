### Linear Combinations

> The linear combination of the vectors  $ùêØ_1,ùêØ_2,‚Ä¶,ùêØ_ùëõ$ with scalars  $ùëê_1,ùëê_2,‚Ä¶,ùëê_ùëõ$ is the vector
> 
> $ùëê_1ùêØ_1+ùëê_2ùêØ_2+‚Ä¶+ùëê_ùëõùêØ_ùëõ.$
> 
> The scalars $ùëê_1,ùëê_2,‚Ä¶,ùëê_ùëõ$ are called the weights of the linear combination.

#### Scalar multiplication

```python
v1 = vector([1,2])
c1 = 2

v1xc1 = c1*v1

plots = v1.plot(color='blue') + v1xc1.plot(color='lightblue', linestyle=':') 
show(plots, figsize=4)
```

#### Vector addition

```python
v1 = vector([1,2])
v2 = vector([1,0])

v1_v2 = v1 + v2

plots = v1.plot(color='blue') + v2.plot(color='blue') + v1_v2.plot(color='red')
show(plots, figsize=4)
```

#### Scalar multiplication and Vector Addition "Combined"

Linear Combination = both Scalar multiplication and Vector Addition 

```python
v1 = vector([1,2])
c1 = 2

v2 = vector([1,0])
c2 = 1

lin_combination = c1*v1 + c2*v2

plots = \
    v1.plot(color='blue') + \
    (c1*v1).plot(color='lightblue', linestyle=':') + \
    v2.plot(color='blue') + \
    lin_combination.plot(color='red')

show(plots, figsize=4)
```
